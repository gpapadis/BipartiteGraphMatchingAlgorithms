```python
import pandas as pd

df = pd.read_csv('../data/schemaBasedSyntacticWeightsResults.csv', sep='|')
df
```

```python
methods = ['CNC', 'RSR', 'RCA', 'BAH', 'BMC', 'EXC', 'KRC',  'UMC']
```

```python
for method in methods:
    df[method + ' First'] = 0
for method in methods:
    df[method + ' Second'] = 0
for method in methods:
    df[method + ' Delta'] = 0
```

```python
for index, row in df.iterrows():
    allF1s = set()
    for method in methods:
        allF1s.add(row[method + ' F-Measure'])
         
    lF1s = list(allF1s)
    lF1s.sort()
    first = lF1s[-1]
    second = lF1s[-2]
              
    for method in methods:
        if (row[method + ' F-Measure'] == first):
            df.loc[index, method + ' First'] = 1
            df.loc[index, method + ' Delta'] = 1 - (second / first)
        elif (row[method + ' F-Measure'] == second):
            df.loc[index, method + ' Second'] = 1
df
```

```python
blc = df[df['Duplicates Type'] == 'Balanced']
for method in methods:
    print('BLC', method, ' First', len(blc[blc[method + ' First'] == 1 ]))
    print('BLC', method, ' Delta' , round(blc[blc[method + ' First']==1][method + ' Delta'].mean()*100, 2))
    print('BLC', method, ' Second', len(blc[blc[method + ' Second'] == 1 ]))
```

```python
osd = df[df['Duplicates Type'] == 'One-sided']
for method in methods:
    print('OSD', method, ' First', len(osd[osd[method + ' First'] == 1 ]))
    print('OSD', method, ' Delta' , round(osd[osd[method + ' First']==1][method + ' Delta'].mean()*100, 2))
    print('OSD', method, ' Second', len(osd[osd[method + ' Second'] == 1 ]))
```

```python
scr = df[df['Duplicates Type'] == 'Scarce']
for method in methods:
    print('SCR', method, ' First', len(scr[scr[method + ' First'] == 1 ]))
    print('SCR', method, ' Delta' , round(scr[scr[method + ' First']==1][method + ' Delta'].mean()*100, 2))
    print('SCR', method, ' Second', len(scr[scr[method + ' Second'] == 1 ]))
```

```python
for method in methods:
    print('OVL', method, ' First', len(df[df[method + ' First'] == 1 ]))
    print('OVL', method, ' Delta' , round(df[df[method + ' First']==1][method + ' Delta'].mean()*100, 2))
    print('OVL', method, ' Second', len(df[df[method + ' Second'] == 1 ]))
```

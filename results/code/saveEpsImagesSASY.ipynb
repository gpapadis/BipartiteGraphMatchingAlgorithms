```python
import pandas as pd

main_dir = 'C:\\Users\\Georgios\\Downloads\\clustering4ccerRevision\\journal\\new\\'
df = pd.read_excel(main_dir + 'schemaAgnosticAnalysisClean.xlsx')
df
```

```python
precision = pd.DataFrame()
for column in df:
    if (0 < column.find('Precision')):
            precision[column] = df[column]
```

```python
recall = pd.DataFrame()
for column in df:
    if (0 < column.find('Recall')):
            recall[column] = df[column]
```

```python
fMeasures = pd.DataFrame()
for column in df:
    if (0 < column.find('F-Measure')):
            fMeasures[column] = df[column]
```

```python
fStarMeasures = pd.DataFrame()
for column in df:
    if (0 < column.find('F-Measure')):
            fStarMeasures[column] = df[column]/(2-df[column])
```

```python
methods = ['CNC', 'RSR', 'RCA', 'BAH', 'BMC', 'EXC', 'KRC',  'UMC']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(font_scale = 1.4)
sns.set_style("white")
```

```python
precision.columns = methods
sns.boxplot(data = precision,
            showmeans=True,
            meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                       "markersize":"10"});
plt.savefig(main_dir + 'saPrecision.eps', format='eps')
```

```python
recall.columns = methods
sns.boxplot(data = recall,
            showmeans=True,
            meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                       "markersize":"10"});
plt.savefig(main_dir + 'saRecall.eps', format='eps')
```

```python
fMeasures.columns = methods
sns.boxplot(data = fMeasures,
            showmeans=True,
            meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                       "markersize":"10"});
plt.savefig(main_dir + 'saF1.eps', format='eps')
```

```python
fStarMeasures.columns = methods
sns.boxplot(data = fStarMeasures,
            showmeans=True,
            meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                       "markersize":"10"});
plt.savefig(main_dir + 'saFstar.eps', format='eps')
```

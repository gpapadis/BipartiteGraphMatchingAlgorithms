```python
import pandas as pd

df = pd.read_csv('../data/allResults.csv', sep='|')
df
```

```python
methods = ['CNC', 'RSR', 'RCA', 'BAH', 'BMC', 'EXC', 'KRC',  'UMC']
```

```python
precision = pd.DataFrame()
for column in df:
    if (0 < column.find('Precision')):
            precision[column] = df[column]
            
precision.columns = methods
```

```python
recall = pd.DataFrame()
for column in df:
    if (0 < column.find('Recall')):
            recall[column] = df[column]

recall.columns = methods
```

```python
fMeasures = pd.DataFrame()
for column in df:
    if (0 < column.find('F-Measure')):
            fMeasures[column] = df[column]
            
fMeasures.columns = methods
```

```python
threshold = pd.DataFrame()
for column in df:
    if (0 < column.find('Threshold')):
            threshold[column] = df[column]

threshold.columns = methods
```

```python
precision.describe().round(3)
```

```python
recall.describe().round(3)
```

```python
fMeasures.describe().round(3)
```

```python
fStarMeasures = pd.DataFrame()
for column in df:
    if (0 < column.find('F-Measure')):
            fStarMeasures[column] = df[column]/(2-df[column])
```

```python
fStarMeasures.describe().round(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(font_scale = 1.4)
sns.set_style("white")
```

```python
precision.columns = methods
p = sns.boxplot(data = precision,
                showmeans=True,
                meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"black",
                           "markersize":"10"})
p.set_title("Precision", fontsize = 30)
# plt.savefig('allPrecision.eps', format='eps')
```

```python
recall.columns = methods
p = sns.boxplot(data = recall,
                showmeans=True,
                meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"black",
                           "markersize":"10"})
p.set_title("Recall", fontsize = 30)
# plt.savefig('allRecall.eps', format='eps')
```

```python
fMeasures.columns = methods
p = sns.boxplot(data = fMeasures,
                showmeans=True,
                meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                       "markersize":"10"})
p.set_title("F1", fontsize = 30)
# plt.savefig('allF1.eps', format='eps')
```

```python
fStarMeasures.columns = methods
p = sns.boxplot(data = fStarMeasures,
                showmeans=True,
                meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"black",
                           "markersize":"10"})
p.set_title("F*", fontsize = 30)
# plt.savefig('allFstar.eps', format='eps')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from autorank import autorank, create_report, plot_stats, latex_table
```

```python
pd.set_option('display.max_columns', 9)
data = pd.DataFrame()

data_csv = pd.read_csv('../data/precisionCriticalDifference.csv')

grouped_df = data_csv.groupby('classifier_name')['accuracy'].apply(list)
for clustering in grouped_df.keys():
    data[clustering] = grouped_df[clustering]

res = autorank(data, alpha=0.05, verbose=False)

plot_stats(res)
plt.title("Precision", fontsize = 30)
# plt.savefig('cdPrecision.eps', format='eps')
```

```python
pd.set_option('display.max_columns', 9)
data = pd.DataFrame()

data_csv = pd.read_csv('../data/recallCriticalDifference.csv')

grouped_df = data_csv.groupby('classifier_name')['accuracy'].apply(list)
for clustering in grouped_df.keys():
    data[clustering] = grouped_df[clustering]

res = autorank(data, alpha=0.05, verbose=False)

plot_stats(res)
plt.title("Recall", fontsize = 30)
plt.xlabel("Recall", fontsize = 30)
# plt.savefig('cdRecall.eps', format='eps')
```

```python
from scipy.stats.stats import pearsonr

methods = ['CNC', 'RSR', 'RCA', 'BAH', 'BMC', 'EXC', 'KRC',  'UMC']
```

```python
import pandas as pd


df = pd.read_csv('../data/schemaBasedSyntacticWeightsResults.csv', sep='|')
df
```

```python
df['Normalized Size'] = df['Valid Pairs']/df['BF Pairs']
```

```python
for method in methods:
    r, _ = pearsonr(df[method + ' Threshold'], df['Normalized Size'])
    print(method, round(r, 2))
```

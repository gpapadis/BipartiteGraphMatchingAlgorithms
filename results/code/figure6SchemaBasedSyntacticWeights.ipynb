```python
import pandas as pd

df = pd.read_csv('../data/schemaBasedSyntacticWeightsResults.csv', sep='|')
df
```

```python
precision = pd.DataFrame()
for column in df:
    if (0 < column.find('Precision')):
            precision[column] = df[column]
```

```python
recall = pd.DataFrame()
for column in df:
    if (0 < column.find('Recall')):
            recall[column] = df[column]
```

```python
fMeasures = pd.DataFrame()
for column in df:
    if (0 < column.find('F-Measure')):
            fMeasures[column] = df[column]
```

```python
fStarMeasures = pd.DataFrame()
for column in df:
    if (0 < column.find('F-Measure')):
            fStarMeasures[column] = df[column]/(2-df[column])
```

```python
methods = ['CNC', 'RSR', 'RCA', 'BAH', 'BMC', 'EXC', 'KRC',  'UMC']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(font_scale = 1.4)
sns.set_style("white")
```

```python
precision.columns = methods
p = sns.boxplot(data = precision,
                showmeans=True,
                meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"black",
                           "markersize":"10"})
p.set_title("Precision", fontsize = 30)
# plt.savefig('sbsyPrecision.eps', format='eps')
```

```python
recall.columns = methods
p = sns.boxplot(data = recall,
                showmeans=True,
                meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"black",
                           "markersize":"10"})
p.set_title("Recall", fontsize = 30)
# plt.savefig('sbsyRecall.eps', format='eps')
```

```python
fMeasures.columns = methods
p = sns.boxplot(data = fMeasures,
                showmeans=True,
                meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"black",
                           "markersize":"10"})
p.set_title("F1", fontsize = 30)
# plt.savefig('sbsyF1.eps', format='eps')
```

```python
fStarMeasures.columns = methods
p = sns.boxplot(data = fStarMeasures,
                showmeans=True,
                meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"black",
                           "markersize":"10"})
p.set_title("F*", fontsize = 30)
# plt.savefig('sbsyFstar.eps', format='eps')
```
